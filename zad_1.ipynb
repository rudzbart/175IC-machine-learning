```
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data = {
    'size': ['L', 'M','XL','L','L','S','XS','S','XL','XXL','L','L','L','L','L','M','S','M','M','XXL'],
    'color': ['green','red','blue','black','white','white','black','black','black','black','brown','yellow','white','blue','green','yellow','black','black','white','brown'],
    'gender': ['female','male','male','male','female','female','male','female','female','female','female','female','female','male','male','male','female','female','male','male'],
    'price': [149.0,79.0,30.0,45.0,59.99,29.99,18.99,149.0,129.0,299.0,129.99,29.99,39.99,50.0,59.99,29.99,80.0,130.0,179.0,229.0],
    'weight': [380,410,400,430,290,325,300,400,425,380,375,300,375,400,290,295,310,320,325,400],
    'bought': ['no','yes','yes','yes','no','no','yes','no','no','no','no','yes','yes','yes','no','no','yes','no','no','no']
}

df_raw = pd.DataFrame(data=data)

print(df_raw)

df = df_raw.copy()

for col in ['size', 'color', 'gender', 'bought']:
 df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float64')

df.info()
df.describe(include=['category']).T
df

le = LabelEncoder()
le.fit_transform(df['bought'])

df['bought'] = le.fit_transform(df['bought'])
df

df = pd.get_dummies(data=df, drop_first=True)

scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df_raw
```
