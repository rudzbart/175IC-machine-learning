```
import numpy as np
import pandas as pd
import plotly.express as px
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data, _ = make_blobs(n_samples=1000, centers=4, cluster_std=1.4, random_state=33)
df = pd.DataFrame(data, columns=['x1','x2'])
px.scatter(df, 'x1', 'x2')

cluster = DBSCAN(eps=0.5, min_samples=9)
cluster.fit(data)

df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', color_continuous_midpoint=0)

df['cluster'] = cluster.labels_
px.scatter(df, 'x1', 'x2', 'cluster', template='plotly_dark', color_continuous_midpoint=0)

plt.scatter(df['x1'], df['x2'], s=8)

#KMEANS
for i in range(1,5):
    plt.figure(figsize=(5, 3))
    y_pred = KMeans(n_clusters=i, random_state=420).fit_predict(df)
plt.scatter(df['x1'], df['x2'], c=y_pred, s=8)
plt.show();

results = pd.DataFrame([df.index,y_pred]).T
print('DBSCAN\n')
print(df['cluster'].value_counts())
print('KMEANS\n')
print(results[1].value_counts())


```
